```python
import pandas as pd

train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
test_df.loc[:, "target"] = -1
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
data = pd.concat((train_df, test_df)).reset_index(drop=True)
```

```python
from sklearn.preprocessing import LabelEncoder

features = [x for x in data.columns if x not in ["id", "target"]]


for feat in features:
    lbl_enc = LabelEncoder()
    tmp_col = data[feat].fillna("NONE").astype(str).values
    new_col = lbl_enc.fit_transform(tmp_col)
    data.loc[:, feat] = new_col
```

```python
train_df = data.loc[data.target != -1].reset_index(drop=True)
test_df = data.loc[data.target == -1].reset_index(drop=True)
```
